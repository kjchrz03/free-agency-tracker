### Split Team Data

```python

split_goalies= pd.read_csv(r"C:\Users\kjchr\Desktop\hockey\new_hockey_env\split_goalies.csv")
split_goalies=split_goalies[['Player', 'Team','GP']].sort_values(by=['Player', 'GP'], ascending=[True, False])

split_players = pd.read_csv(r"C:\Users\kjchr\Desktop\hockey\new_hockey_env\split_times_2026.csv")
split_players= split_players[['Player', 'Team', 'GP']].sort_values(by=['Player', 'GP'], ascending=[True, False])

all_split_players = pd.concat([split_players, split_goalies])
all_split_players.head()
```

```python
#first, ID players who had been traded during the season
traded_ufas=all_split_players[['Player', 'Team', 'GP']].sort_values(by=['Player', 'GP'], ascending=[True, False])
traded_ufas = traded_ufas[traded_ufas["Player"].duplicated(keep=False)]
traded_ufas.head(10)
```

```python
#Id the team that they ended with

ending_teams = traded_ufas.drop_duplicates(
    subset=["Player"],
    keep="last"
)

ending_teams = ending_teams.sort_values(by='Player').reset_index(drop=True)

ending_teams = ending_teams.rename(columns = {'Player':'player_name', 'Team':'tri_code_2526'}).drop(columns= 'GP')
```

### Off season trades

```python
off_season_trade = all_split_players[['Player', 'Team']].drop_duplicates(keep=False).reset_index(drop=True)
off_season_trade['Team'] = np.where(off_season_trade['Team']=='T.B', 'TBL', off_season_trade['Team'])
off_season_trade['Team'] = np.where(off_season_trade['Team']=='N.J', 'NJD', off_season_trade['Team'])
off_season_trade['Team'] = np.where(off_season_trade['Team']=='L.A', 'LAK', off_season_trade['Team'])
off_season_trade['Team'] = np.where(off_season_trade['Team']=='S.J', 'SJS', off_season_trade['Team'])
```

```python
players[players['player_name'].str.contains('markstr', case=False, na=False)]
```

```python
off_season_trade=off_season_trade.rename(columns = {'Player':'player_name'})
off_season_trade.head()
```

```python
#These players have active contracts and currently sit on a roster
non_fa_players = players[players['team_id_2526'].notna()]
non_fa_players.head()
```

```python
off_season_trade =off_season_trade.merge(non_fa_players, on='player_name', how = 'inner')
x_cols = [c for c in off_season_trade.columns if c.endswith("_x")]

for x in x_cols:
    base = x[:-2]              # remove "_x"
    y = f"{base}_y"

    if y in off_season_trade.columns:
        off_season_trade[base] = off_season_trade[x].combine_first(off_season_trade[y])
        off_season_trade.drop(columns=[x, y], inplace=True)
off_season_trade.head()
```

```python
#ID only the players who had been traded in the off season 
off_season_trade['action'] = np.where(off_season_trade['Team']!=off_season_trade['tri_code_2526'], "off-season trade", "")
off_season_trade = off_season_trade[off_season_trade['action']=='off-season trade'].drop_duplicates()
off_season_trade.head()
```

```python
off_season_trade = off_season_trade.drop(columns = 'Team')
```

### Free Agents

```python
# players who do not currently sit on a roster
free_agents = players[players['team_id_2526'].isna()]
free_agents.head()
```

```python
#Updating the 25-26 teams for players who were traded in-season and still went to free agency

free_agents = free_agents.merge(ending_teams, on='player_name', how='left')
x_cols = [c for c in free_agents.columns if c.endswith("_x")]

for x in x_cols:
    base = x[:-2]              # remove "_x"
    y = f"{base}_y"

    if y in free_agents.columns:
        free_agents[base] = free_agents[x].combine_first(free_agents[y])
        free_agents.drop(columns=[x, y], inplace=True)
free_agents.info()
```

```python
free_agents[free_agents['player_name'].str.contains('marner', case=False, na=False)]
```

```python
free_agents["player_id"] = free_agents["player_id"].astype('Int64').astype(int)
free_agents.head()
```

```python
free_agents["action"] = np.where(free_agents['tri_code_2526'].notna(), "traded", "free agent")
free_agents.head()
```

```python

free_agents['tri_code_2526'] = np.where(free_agents['tri_code_2526']=='T.B', 'TBL', free_agents['tri_code_2526'])
free_agents['tri_code_2526'] = np.where(free_agents['tri_code_2526']=='N.J', 'NJD', free_agents['tri_code_2526']) 
free_agents['tri_code_2526'] = np.where(free_agents['tri_code_2526']=='L.A', 'LAK', free_agents['tri_code_2526'])
free_agents['tri_code_2526'] = np.where(free_agents['tri_code_2526']=='S.J', 'SJS', free_agents['tri_code_2526'])
```

```python
for index, row in free_agents.iterrows():
    if row['action'] == "free agent":
        free_agents.at[index, 'team_id_2526'] = row['team_id_2425']
        free_agents.at[index, 'team_name_2526'] = row['team_name_2425']
        free_agents.at[index, 'tri_code_2526'] = row['tri_code_2425']
        
free_agents.head(10)
```

```python

free_agents = free_agents.merge(
        teams,
       on='tri_code_2526',
        how="left",
        validate="m:1",
        indicator=True,
    )

x_cols = [c for c in free_agents.columns if c.endswith("_x")]

for x in x_cols:
    base = x[:-2]              # remove "_x"
    y = f"{base}_y"

    if y in free_agents.columns:
        free_agents[base] = free_agents[x].combine_first(free_agents[y])
        free_agents.drop(columns=[x, y], inplace=True)
free_agents.head()
```

```python
free_agents = free_agents.drop(columns = '_merge')
```

```python
off_season = pd.DataFrame()
player_url = player_movement["player_url"].tolist()


def process_player_batch(player_batch):
    batch_players = []
    session = requests.Session()

    for item in player_batch:
        player_url = item["player_url"]

        while True:
            response = session.get(player_url)

            if response.status_code == 200:
                break

            if response.status_code == 429:
                wait = int(response.headers.get("Retry-After", 5))
                print(f"Rate limited. Waiting {wait} seconds...")
                time.sleep(wait)
                continue

            print(f"Request failed with status code {response.status_code}")
            response = None
            break

        if response is None:
            continue

        json_data = json.loads(response.text)

        player = pd.json_normalize(json_data).reindex(
            columns=[
                "playerId",
                "isActive",
                "currentTeamId",
                "currentTeamAbbrev",
            ]
        )

        player[["currentTeamId", "currentTeamAbbrev"]] = player[
            ["currentTeamId", "currentTeamAbbrev"]
        ].where(
            player[["currentTeamId", "currentTeamAbbrev"]].notna(),
            None
        )

        batch_players.append(player)

        time.sleep(0.2)

    if batch_players:
        return pd.concat(batch_players, ignore_index=True)

    return pd.DataFrame(
        columns=[
            "playerId",
            "isActive",
            "currentTeamId",
            "currentTeamAbbrev",
        ]
    )


player_batch = [{"player_url": url} for url in player_url]

batch_size = 100

for i in range(0, len(player_batch), batch_size):
    batch = player_batch[i:i + batch_size]
    batch_players = process_player_batch(batch)

    if not batch_players.empty:
        off_season = pd.concat(
            [off_season, batch_players],
            ignore_index=True
        )

off_season.reset_index(drop=True, inplace=True)



export_path = r"C:\Users\kjchr\Desktop\hockey\new_hockey_env\API data\off_season2.csv"
off_season.to_csv(export_path, index=False)
```
